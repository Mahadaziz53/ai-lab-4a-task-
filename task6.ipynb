Task 6
: BFS without Queue & without Node

```python
def bfs(graph, start):
    visited = []
    stack = [start]  

    while stack:
        node = stack.pop(0)  
        if node not in visited:
            visited.append(node)
            stack.extend(graph[node])  

    return visited

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

print("BFS without Queue & Node:", bfs(graph, 'A'))
```

BFS with Queue & Node

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []

def bfs_with_queue(root):
    queue = deque([root])
    visited = []

    while queue:
        node = queue.popleft()
        visited.append(node.value)
        queue.extend(node.children)

    return visited


root = Node(1)
root.children = [Node(2), Node(3), Node(4)]
root.children[0].children = [Node(5), Node(6)]

print("BFS with Queue & Node:", bfs_with_queue(root))
```
