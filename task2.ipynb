Today's project is actually a very common interview question, which revolves around a childhood
counting game called Fizz Buzz.

```python
class FizzBuzz:

    def __init__(self, limit):
        self.limit = limit

    def generate(self):
        for number in range(1, self.limit + 1):

            if number % 15 == 0:
                print("Fizz Buzz")

            elif number % 3 == 0:
                print("Fizz")

            elif number % 5 == 0:
                print("Buzz")

            else:
                print(number)


def main():
    try:
        user_limit = int(input("Enter the limit for Fizz Buzz: "))
        
        if user_limit <= 0:
            print("Please enter a positive number.")
            return
        
        game = FizzBuzz(user_limit)
        game.generate()

    except ValueError:
        print("Invalid input! Please enter a valid integer.")


if __name__ == "__main__":
    main()
```
