Lab 4 Task
1. Code for LUHN Algorithm
2. Write a python program to remove punctuations from the given string?
3. Write a python program to sort the sentence in alphabetical order?

```python
import string

# ----- Task 1: LUHN Algorithm -----
def luhn_check(card_number):
    digits = [int(d) for d in str(card_number)][::-1]
    total = 0

    for i, d in enumerate(digits):
        if i % 2 == 1:
            d *= 2
            if d > 9:
                d -= 9
        total += d

    return total % 10 == 0

# ----- Task 2: Remove Punctuations -----
def remove_punctuations(text):
    return "".join([ch for ch in text if ch not in string.punctuation])

# ----- Task 3: Sort Sentence Alphabetically -----
def sort_sentence(sentence):
    words = sentence.split()
    words.sort()
    return " ".join(words)

# ----- Menu -----
def main():
    while True:
        print("\n--- Lab 4 Menu ---")
        print("1. LUHN Algorithm (Card Validation)")
        print("2. Remove Punctuations from String")
        print("3. Sort Sentence Alphabetically")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            card = input("Enter card number: ")
            if luhn_check(card):
                print("Valid Card ")
            else:
                print("Invalid Card ")

        elif choice == "2":
            text = input("Enter text: ")
            print("Text without punctuation:", remove_punctuations(text))

        elif choice == "3":
            sentence = input("Enter a sentence: ")
            print("Sorted Sentence:", sort_sentence(sentence))

        elif choice == "4":
            print("Exiting Lab 4 ")
            break

        else:
            print("Invalid choice! Please enter 1-4.")


if __name__ == "__main__":
    main()
```
