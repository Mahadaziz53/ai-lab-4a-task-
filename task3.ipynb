Lab 3 Task:
• Model-Based Reflex Agent
• This agent not only checks the current temperature but also remembers the previous
action to avoid turning the heater on or off unnecessarily

```python


previous_action = None  

while True:
    temp = input("Enter temperature (or 'exit' to stop): ")
    
    if temp.lower() == "exit":
        print("System stopped ")
        break

    try:
        temp = float(temp)
    except:
        print("Invalid input! Enter a number.")
        continue

    if temp < 20:
        action = "Heater ON"
    elif temp > 25:
        action = "Heater OFF"
    else:
        action = "No Action"

    if action == previous_action:
        print(f"Temperature: {temp}°C → No change (Still {previous_action})")
    else:
        print(f"Temperature: {temp}°C → {action}")
        previous_action = action
```
