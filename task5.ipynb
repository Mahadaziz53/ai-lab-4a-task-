1. DFS with Stack & Node

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def dfs_stack(root):
    stack = [root]
    visited = []

    while stack:
        node = stack.pop()
        if node:
            visited.append(node.value)
            stack.append(node.right)
            stack.append(node.left)

    return visited


root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print("DFS using Stack:", dfs_stack(root))
```

2. Research about "Inorder, Preorder, Postorder" and implement in DFS

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Preorder Traversal
def preorder(root):
    if root is None:
        return []
    return [root.value] + preorder(root.left) + preorder(root.right)

# Inorder Traversal
def inorder(root):
    if root is None:
        return []
    return inorder(root.left) + [root.value] + inorder(root.right)

# Postorder Traversal
def postorder(root):
    if root is None:
        return []
    return postorder(root.left) + postorder(root.right) + [root.value]

# ---- Example Tree ----
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print("Preorder:", preorder(root))
print("Inorder:", inorder(root))
print("Postorder:", postorder(root))
```
